```python
from sklearn.datasets import make_regression
import numpy as np

X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise = 80, random_state= 13)
```

```python
X.shape
```

```python
import matplotlib.pyplot as plt

plt.scatter(X,y)
```

```python
#Let's apply OLS to find cofficient(m) & intercept

from sklearn.linear_model import LinearRegression
reg = LinearRegression()

#fitting model
reg.fit(X,y)
```

```python
print('Coefficient (m) : ', reg.coef_)
print('Intercept (b) : ', reg.intercept_)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color = 'red')
```

```python
# Lets apply Gradient Descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0
m = 78.35063668
b = 0
y_pred = ((m * X.ravel()) + b).reshape(4)
```

```python
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color = 'red', label = 'OLS')
plt.plot(X,y_pred, color = 'blue', label = 'b = {}'.format(b))
plt.legend()
```

# Iterating the below steps

```python
loss_slope = -2 * np.sum(y - (m*X.ravel()) - b)
# b(new) = b(old) - learning_rate *(slope)
#setting learning rate = 0.1
lr = 0.1
step_size = lr * loss_slope
```

```python
#Calculating new intercept
b = b - step_size
```

```python
y_pred1 = (m * X.ravel()) + b
```

```python

y_pred1 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred1,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python
y_pred2 = ((78.35 * X) + b).reshape(4)

plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred2,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b)
loss_slope
```

```python
step_size = loss_slope*lr
step_size
```

```python
b = b - step_size
b
```

```python

y_pred3 = ((78.35 * X) + b).reshape(4)

plt.figure(figsize=(10,10))
plt.scatter(X,y)
plt.plot(X,reg.predict(X),color='red',label='OLS')
plt.plot(X,y_pred3,color='#00a65a',label='b = {}'.format(b))
plt.plot(X,y_pred2,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred1,color='#A3E4D7',label='b = {}'.format(b))
plt.plot(X,y_pred,color='#A3E4D7',label='b = 0')
plt.legend()
plt.show()
```

# Code using For loop to make iteration simple

```python
m = 78.35063668
b = -100            #(starting point)
learning_rate = 0.1

# l_slope = -2 * np.sum(y - (m * X.ravel()) - b)
# step_size = l_slope * learning_rate

epochs = 100   #epochs is number of iterations to be performed

for i in range(epochs) :

  l_slope = -2 * np.sum(y - (m * X.ravel()) - b)
  step_size = l_slope * learning_rate

  b = b - step_size

  y_pred = (m * X.ravel()) + b

  plt.plot(X, y_pred)

plt.scatter(X, reg.predict(X), color = 'red')
```
