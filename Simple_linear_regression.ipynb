```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Importing placement dataset
df = pd.read_csv('/content/placement.csv')
df.head()
# df.info()
```

```python
#plotting the data to understand the distribution
plt.scatter(df['cgpa'], df['package'])
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
plt.tight_layout()
plt.show()
```

```python
x = df.iloc[:,0:1]   #here written col as 0:1, representing datafame(2D) of 1 coln bcuz sklearn takes input as 2D argument
x
```

```python
y = df.iloc[:,-1]
y
```

```python
#Split placement data into train(80% data) & test(20% data) data

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=2)
```

```python
#fitting simple linear regression model onto the data OR making my model learn from the data & draw a prediction
#(best fit line ) onto the data

from sklearn.linear_model import LinearRegression

lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test.iloc[0].values.reshape(1,1))
```

```python
#to check whether our predict model has predicted it correct or not
print('x test : ', x_test)
print('y test : ', y_test)
```

```python
#Plotting training data & predicted best-fit-line from the model

plt.scatter(df['cgpa'],df['package'], label = 'Provided Data')
plt.plot(x_train, lr.predict(x_train), color = 'red', label = 'Predicted best-fit-line')
plt.xlabel('CGPA')
plt.ylabel('Package (in LPA)')
plt.legend()
plt.tight_layout()
plt.show()
```
