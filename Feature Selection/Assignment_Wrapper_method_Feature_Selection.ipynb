# Applying SFS Backward elimination

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
#baseline model
from sklearn.metrics import accuracy_score

reg = LogisticRegression(max_iter=1000)

print('train data : ', np.mean(cross_val_score(reg, X_train, y_train, scoring='accuracy', cv=5)))
print('test data : ', np.mean(cross_val_score(reg, X_test, y_test, scoring='accuracy', cv=5)))
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
lr = LogisticRegression(max_iter=1000)

sel1 = SFS(lr, k_features='best', forward=False, floating=False, scoring='accuracy', cv=5, n_jobs=-1)
model1 = sel1.fit(X_train, y_train)
```

# SFS Forward Selection

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=3)
X_train.shape
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from mlxtend.feature_selection import SequentialFeatureSelector as SFS
lr = LogisticRegression(max_iter=1000)

sel2 = SFS(lr, k_features='best', forward=True, floating=False, scoring='accuracy', cv=5, n_jobs=-1)
model2 = sel2.fit(X_train, y_train)
```

```python
metric_df2 = pd.DataFrame.from_dict(model2.get_metric_dict()).T
metric_df2.sort_values('avg_score', ascending=False)
```

```python
X_train = sel2.transform(X_train)
X_test = sel2.transform(X_test)
```

```python
#Checking accuracy after applying Feature selection
lr = LogisticRegression()
print('On training data : ', np.mean(cross_val_score(lr, X_train, y_train, scoring='accuracy', cv=5, n_jobs=-1)))
print('On test data : ', np.mean(cross_val_score(lr, X_test, y_test, scoring='accuracy', cv=5, n_jobs=-1)))
```
