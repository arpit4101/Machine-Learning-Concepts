# Creating our own class for KNN from scratch

```python
import numpy as np
from collections import Counter
class Knn:

  def __init__(self,n_neighbors=5):

    self.k = n_neighbors
    self.X_train = None
    self.y_train = None

  def fit(self, X_train, y_train):

    self.X_train = X_train
    self.y_train = y_train

  def predict(self, X_test):

    y_pred = []

    for i in X_test:
      #calculate distances with each training point
      distances = []

      for j in X_train:

        distances.append(self.calculate_distance(i, j))

      sorted_nearest_neighbors = sorted(list(enumerate(distances)), key=lambda x: x[1], reverse=False)[0:self.k]
      label = self.majority_count(sorted_nearest_neighbors)
      y_pred.append(label)

    return np.array(y_pred)


  def majority_count(self, nearest_neighbors):

    votes = []

    for i in nearest_neighbors:
      # print(y_train[i[0]])
      votes.append(y_train[i[0]])
    # print(votes)
    counts = Counter(votes)
    return counts.most_common()[0][0]


  def calculate_distance(self,point_A, point_B):

    return np.linalg.norm(point_A - point_B)
```

```python
knn = Knn()
knn.fit(X_train,y_train)
# knn.predict(X_test)
print('accuracy_score', accuracy_score(y_test,knn.predict(X_test)))
```
